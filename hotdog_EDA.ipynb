```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/ashleynguyen04/DS-4002/refs/heads/main/hot-dog-survey-data/fa25/Is%20hotdog%20a%20sandwich_%20(Responses).xlsx"
df = pd.read_excel(url)
df.head()
```

```python
response_col = "Is hotdog a sandwich?"
year_col = "What year are you in?"

counts = df.groupby([response_col, year_col]).size().unstack(fill_value=0)

counts.plot(kind="bar", figsize=(8,5))

plt.title("Hotdog Sandwich Responses Divided by Academic Year")
plt.xlabel("Response (Yes/No)")
plt.ylabel("Count")
plt.legend(title="Academic Year")
plt.show()
```
