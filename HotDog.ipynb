```python
# Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Get the data
data = pd.read_csv('/content/hot_dog_survey_spring25.xlsx - Form Responses 1.csv')
data.head()
```

```python
# Get rid of columns we don't care about
results = []
letter = "y"

results.append(data.keys())

for index, row in data.iterrows():
  if letter.lower() in row['Are you currently enrolled as an undergraduate or graduate student at the University of Virginia?'].lower():
    results.append(row)


print(type(results))
```

```python
# Making the results into an array so easier to deal with
res = np.array(results)
print(res)
```

```python
# Deleting columns we don't care about
clean = np.delete(res, [0, 2, 3, 7, 8], 1)

# Renaming columns
clean[0][0] = "response"
clean[0][1] = "major"
clean[0][2] = "birthplace"
clean[0][3] = "year"
print(clean)
```

```python
# change yesses and nos
for i in range(1, len(clean)):
  if letter.lower() in clean[i][0]:
    clean[i][0] = "yes"
  else:
    clean[i][0] = "no"

print(clean)

# change virginia
```

```python
# Changing birthplaces
s = "Virginia"
s2 = "VA"
for i in range(1, len(clean)):
  if (s.lower() or s2.lower()) in str(clean[i][2]).lower():
    clean[i][2] = "virginia"
  else:
    clean[i][2] = "other"

print(clean)
```

```python
# Making plots
clean_df = pd.DataFrame(clean[1:], columns = clean[0])
#print(clean_df)
# Draw a nested barplot by species and sex

answer = clean_df['response']
bp = clean_df['birthplace']
yr = clean_df['year']
maj = clean_df['major']

from operator import le

cross1 = pd.crosstab(answer, bp)
cross2 = pd.crosstab(answer, yr)

boop = cross1.plot.bar(rot=0, stacked=True, color=sns.husl_palette(len(cross1.columns)))
plt.xlabel('Response')
legend_handles, _ = boop.get_legend_handles_labels()
boop.legend()

plt.ylabel('Count')
plt.show()

boop2 = cross2.plot.bar(rot=0, stacked=True, color=sns.husl_palette(len(cross2.columns)))
plt.xlabel('Response')
legend_handles, _ = boop2.get_legend_handles_labels()
boop2.legend()

plt.ylabel('Count')
plt.show()
```
