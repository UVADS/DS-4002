### Is a Hotdog a Sandwich Survery Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('DS-4002-sp24-survey-results.xlsx')
df
```

```python
# rename column 2 as Sandwich?
# rename column 3 as Rationale?
df.columns = ['Timestamp', 'Sandwich?', 'Rationale?', 'Degree Type?', 'Major?', 'Year', 'Birthplace']
df
```

```python
# remove Degree Type Graduate
df = df[df['Degree Type?'] != 'Graduate']
df
```

```python
# plot number of people Sandwich yes vs no
df['Sandwich?'].value_counts().plot(kind='bar')
```

```python
# group by year, and plot number of people Sandwich yes vs no
# group by year and sandwich status, and plot number of people with different colors for yes and no
df.groupby(['Year', 'Sandwich?'])['Sandwich?'].count().unstack().plot(kind='bar', color=['blue', 'red'])
```
