```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
hotdog_df = pd.read_csv("C:/Users/Govind Molakalapalli/Downloads/Is hotdog a sandwich_ (Responses).csv", encoding='latin1')
hotdog_df.head()
```

```python
hotdog_df['What year are you in?'].value_counts().plot(kind='bar')
```

```python
hotdog_df.groupby(['What year are you in?', 'Is hotdog a sandwich?']).size().unstack().plot(kind='bar')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Hotdog is a Sandwich: Yes/No by Year')
plt.show()
```

```python
years = hotdog_df['What year are you in?'].unique()
fig, axes = plt.subplots(1, len(years), figsize=(5 * len(years), 5))

# Ensure axes is always iterable
if len(years) == 1:
    axes = [axes]

for ax, year in zip(axes, years):
    hotdog_df[hotdog_df['What year are you in?'] == year]['Is hotdog a sandwich?'].value_counts().plot(
        kind='pie', autopct='%1.1f%%', ax=ax)
    ax.set_ylabel('')
    ax.set_title(f'Year {year}')

plt.tight_layout()
fig.set_size_inches(7 * len(years), 7)
plt.show()
```
